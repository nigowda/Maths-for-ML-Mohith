```python
with open('/kaggle/input/glove-global-vectors-for-word-representation/glove.6B.50d.txt') as f:
    embeddings = f.read()
```

```python
embeddings[:500]
```

```python
embeddings_list = embeddings.split('\n')
embeddings_list[:3]
```

```python
data = {}


for emb in embeddings_list[:-1]:
    word = emb.split()[0]
    vec = emb.split()[1:]
    data[word] = np.array(vec, dtype='float')
```
