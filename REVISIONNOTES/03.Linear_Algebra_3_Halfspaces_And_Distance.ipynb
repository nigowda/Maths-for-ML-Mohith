## Shifting 2D lines

**Note:**
here we are working in the co-ordinate geometry space and not working with vectors




**Q. How to shift line by 'a' units to the right?**
- First let's ensure that the line passes through the origing for that we'll make the bias term equal to 0

**Points to remember:**

**Note:** Here origin moving means position of origin with respect to the line
- When shigting a line towards the right the origin moves towards left and if we shift the line to the left then the origin is moving to the right.
- When shifting the line up our origin is moving down and when shifting line down our origin is moving up
- Thus we can say that the origin is moving in the opposite direction to where we shift our line

**Let's see how to represent this mathematically**

Q. When we are asked to shift a line to the right what will we change in the equation of line $w_1x_1+w_2x_2=w_0=0$ ?
- Since the line is passing through origin and we want it to move towards the right let say by 20
- we also know that when we move the line towards right the origin should be on the left side of the line
- so what we'll be doing is we'll be moving origin by 20 units to the left which ensures that the origin is towards the left hand side of the line or we can also say that we are moving the line towards right.
- thus the updated equation of the line is $w_1(x_1-20)+w_2x_2=w_0=0$

**Always think of it as moving the origin and you'll not get confused.**
- So whenever we say that we have sifted the line by 10 units to the right esentially we can say that we have shifted the origin towards left by 10 units.

![](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/047/953/original/Screenshot_2023-09-12_at_6.44.05_PM.png?1694524461)

Q. Now if we want to shif the above line towards left what will we do?
- we'll simply add 20 instead of subtracting which ensures that the origin stays towards the right hand side of the line
- Update equation of line is $w_1(x_1+20)+w_2x_2=w_0=0$

#### Geometric proof

Problem Formulation
- Suppose we have a line (blue) that denotes the decision boundary.
- We kow that vector $w$ (red line) will be perpendicular to this blue line.
- Now, we want to find the distance `d`

image.png

Suppose we choose a random point on blue line: $x_0$

From this random point, we draw a line (yellow) connecting to origin (0, 0)

Let the angle between yellow and red line be $$

<br>

We know the equation of blue line: $w^T x + w_0 = 0$

Therefore point $x_0$ must satisfy this equation, as it is lying on this line: $w^T x_0 + w_0 = 0$

i.e. $w^T x_0 = - w_0$

<br>

Also, we can evaluate the value of `d` using trigonometry as: $cos  = \frac{d}{||x_0||}$
- where $||x_0||$ denotes the length of $x_0$ or it's distance from origin

Therefore, $d = ||x_0|| cos $

image.png

> **What is $cos $ in terms of $x_0$ and $w$?**

We can see $x_0$ and $w$ in terms of vectors.

Recall from last class we derived this to be

$cos  = \frac{x^T_1 . x_2}{||x_1|| . ||x_2||}$

Hence we can say that

$cos  = \frac{w^T . x_0}{||w|| . ||x_0||}$

image.png

Substituting this into equation of d,

$d = ||x_0|| . \frac{w^T . x_0}{||w|| . ||x_0||} = \frac{w^T . x_0}{||w||}$

<br>

Substituting the value of $w^T x_0 = -w_0$

We get: $d = \frac{-w_0}{||w||}$ ... (Equation 1)

image.png

> **Why is the distance negative?**

This sign indicates which side does the origin lies, with respect to the line $w^T.x + w_0 = 0$

This makes sense also. By looking at the position of blue line and red line, we know from last lecture that value of $w_0$ has to be **negative**

##### **Geometric proof**

So, consider we have a point $x_0$.

We need to find the perpendicular distance between $x_0$ and the decision boundary (blue line).

This distance is denoted as `d` (yellow line).

image.png

Let's draw the $\vec{x_0}$ vector (i.e. root at origin, and tip at the point $x_0$)

<br>

We can see that this vector is getting intersected by the decision boundary.

Let the length of each part be `p` and `q`, such that: $||x_0|| = p+q$ ... [Equation 2]

<br>

Also, we just proved above that perpendicular distance `A` between origin and decision boundary is: $A = \frac{-w_o}{||w||}$ .... [Equation 3]

<br>

We label the angle  as shown

This gives us: $cos  = \frac{A}{p}$

or, $p = \frac{A}{cos }$ .... [Equation 4]

<br>

Also, note that the alternate angle will also be , as they are formed by the same line across 2 parallell lines.

This gives us: $cos  = \frac{d}{q}$

or, $d = q . cos $ .... [Equation 5]

<br>

Using equation 2, 4, and 4, we get:

$q = ||x_0|| - p = ||x_0|| - \frac{A}{cos }$


image.png


Therefore substituting `q` : $d = ||x_0| cos  - A$

<br>

Exanding this equation using 1 and 3,

$d = ||x_0|| . (\frac{w^T.x_0}{||w|| . ||x_0||}) - (\frac{-w_0}{||w||})$

<br>

This gives us: $d = \frac{w^T x_0 + w_0}{||w||}$

This relation is our key takeaway.

## Halfspaces

We saw that the line divided the given 2D space into 2 parts one on the left side and other on the right
- Now when we are given a new point we need to find where does the point lie with respect to the line based on that we can tell which class data point does it belong to.

Q. How do we identify this mathematically?
- We are going to use the weight vector
- Points on the right side of the line or along the direction of weight vector we say that they belong to Positive half space
- and points on the left side of the line or opposite to the direction of weight vector we say that they belong to negative half space.

![](https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/048/064/original/Screenshot_2023-09-12_at_11.32.58_PM.png?1694541796)

Let's verify this
- Let's take a scenario where in for the given 2 class classification problem we have a line with equation x=5 seperating them
- Here the weight vector is along the x-axis
- now for points that are on the right side of the line into the equation of line we will get a positive answer
  - Eg. check for point (6,5)
    - if we replace this in our equation of line we will get 6-5
    - which is nothing but 1 which is a positive value
- Now lets check for points that are to the left of our line
  -Eg (2,3)
    - When we substitute it into the equation of line we get 2-5
    - which is -3 which is a negative value
